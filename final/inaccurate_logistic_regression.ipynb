```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.concatenate((np.random.uniform(0,0.5, 500),np.random.uniform(0.5,1, 500),np.random.uniform(0,0.5, 500), np.random.uniform(0.5,1, 500)),axis=0)
x2 = np.concatenate((np.random.uniform(0,0.5, 500),np.random.uniform(0.5,1, 500),np.random.uniform(0.5,1, 500), np.random.uniform(0,0.5, 500)),axis=0)
y = np.append(np.zeros(1000), np.ones(1000))

def scatterValue(x1, x2, Y, color1, color2, marker, label1, label2, s):
    x1_pos = []
    x2_pos = []
    x1_neg = []
    x2_neg = []
    # devide the data by y value
    for i in range(0, len(Y)):
        if Y[i] == 1:
            x1_pos.append(x1[i])
            x2_pos.append(x2[i])
        elif Y[i] == 0:
            x1_neg.append(x1[i])
            x2_neg.append(x2[i])
    # draw the scatter plot
    plt.scatter(x1_pos, x2_pos, c = color1, label = label1, marker=marker, s=s)
    plt.scatter(x1_neg, x2_neg, c = color2, label = label2, marker=marker, s=s)
    plt.xlabel("X_1")
    plt.ylabel("X_2")

scatterValue(x1, x2, y, "r", "b", "+", "class red", "class blue", 20)

plt.legend()
# show the figure
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score, train_test_split

x = np.column_stack((x1, x2))
# split up our data into train and test
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2)

model = LogisticRegression()
score = cross_val_score(model,  xtrain, ytrain, cv=5, scoring='accuracy')
print(score.mean())
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=5, weights='uniform')
score = cross_val_score(model,  xtrain, ytrain, cv=5, scoring='accuracy')
print(score.mean())
```
